```python
import pandas as pd
```

```python
df = pd.read_csv("ncr_ride_bookings.csv")
```

```python
df.describe()
```

```python
df[["Booking Value", "Avg VTAT"]].iloc[:100].plot(kind="scatter", y="Booking Value", x="Avg VTAT")
```

```python
df[["Ride Distance", "Avg CTAT"]].iloc[:200].plot(kind="scatter", y="Ride Distance", x="Avg CTAT")
```

```python
df.groupby("Payment Method")["Date"].count().plot(kind="bar")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df_no_na = df[["Ride Distance", "Avg VTAT"]].dropna()
```

```python
model = LinearRegression()
model.fit(df_no_na[["Avg VTAT"]], df_no_na[["Ride Distance"]])
```
